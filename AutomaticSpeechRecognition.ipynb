```python
import gradio as gr
import os

def transcribe_speech(filepath):
    if filepath is None:
        gr.Warning("No audio found, please retry.")
        return ""
    # Assume asr function is defined elsewhere and handles transcription logic
    output = asr(filepath)
    return output["text"]

# Create Blocks instance
demo = gr.Blocks()

# Define Interfaces within the Blocks context
with demo:
    with gr.Tabs():  # Use Tabs instead of TabbedInterface
        with gr.TabItem("Transcribe Microphone"):
            mic_input = gr.Audio(sources="microphone", type="filepath")
            mic_output = gr.Textbox(label="Transcription", lines=3)
            mic_input.change(fn=transcribe_speech, inputs=mic_input, outputs=mic_output) # Connect function to input
        with gr.TabItem("Transcribe Audio File"):
            file_input = gr.Audio(sources="upload", type="filepath")
            file_output = gr.Textbox(label="Transcription", lines=3)
            file_input.change(fn=transcribe_speech, inputs=file_input, outputs=file_output) # Connect function to input

# Check if PORT1 is set, otherwise use a default port (e.g., 7860)
port = int(os.environ.get('PORT1', 7860))

# Launch the demo
demo.launch(share=True, server_port=port)
```
